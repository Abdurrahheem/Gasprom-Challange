```python
import numpy as np 
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
datapath = Path.cwd().parent / 'data'
```

```python
data_train = np.load(datapath/'data_train.npz', allow_pickle=True, mmap_mode='r')
data_train = data_train['data']

labels_train = np.load(datapath/'labels_train.npz', allow_pickle=True, mmap_mode='r')
labels_train = labels_train['labels']

data_test = np.load(datapath/'data_test_1.npz', allow_pickle=True, mmap_mode='r')
data_test = data_test['data']
```

```python
data_train.shape, labels_train.shape, data_test.shape
```

As we can see axis 0 and 1 exactly match in `data_train` and `data_test`. Let's explote the data in more detail. First, visualize the data through all axis.

```python
plt.figure(figsize=(10,12))

# going forward through Y axis
plt.subplot(3,2,1)
plt.imshow(data_train[:,:,1], cmap='gray')
plt.subplot(3,2,2)
plt.imshow(labels_train[:,:,1])

# going forward through Z axis
plt.subplot(3,2,3)
plt.imshow(data_train[1,:,:], cmap='gray')
plt.subplot(3,2,4)
plt.imshow(labels_train[1,:,:])

# going forward through X axis
plt.subplot(3,2,5)
plt.imshow(data_train[:,1,:], cmap='gray')
plt.subplot(3,2,6)
plt.imshow(labels_train[:,1,:])
```

We could build Unet and conduct segmentation into 6 classes. We can use Y axis as a deepth axis(objects of traning set) in this problem since X and Z dimentions does not change their length in test set but Y dimention does change it's length

```python
for i in range(0, 400, 10):
  plt.subplot(1,2,1)
  plt.imshow(data_train[:,:,i], cmap='gray')
  plt.subplot(1,2,2)
  plt.imshow(labels_train[:,:,i])
  plt.show()
```
